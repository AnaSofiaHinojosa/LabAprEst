```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

data = pd.read_csv('Salarios_minimos.csv')
data.head()
```

## Ciclo para baterias
Para cada grado se muestra el r2 y la gráfica. Modifiqué los datos para que solo aparezcan aquellos datos cuya carga no supera 4.

```python
data = pd.read_csv('bateria_tableta.txt')
data = data[data['carga'] <= 4]
data.head()
```

```python

x = data[['carga']]
y = data['duracion']
lr = LinearRegression()
```

```python
for grado in [1, 3, 5, 7]:
    x_i = np.hstack([x ** d for d in range(1, grado + 1)])
    
    lr.fit(x_i, y)
    print(f"Modelo de grado {grado}: Intercepto = {lr.intercept_}, Coeficientes = {lr.coef_}, r2 = {r2_score(y, lr.predict(x_i))}")

    plt.scatter(x, y, label='Datos originales', color='palevioletred')
    plt.plot(x, lr.predict(x_i), label=f'Modelo grado {grado}', color='darkslategray')
    plt.legend()
    plt.grid()
    plt.show()
```

## Ciclo para salarios

```python
data = pd.read_csv('Salarios_minimos.csv')
x = data[['Periodo']]
y = data['Salario']
lr = LinearRegression()
```
