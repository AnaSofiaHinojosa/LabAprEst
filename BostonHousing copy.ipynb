```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_excel("Boston Housing Dataset 1978.xlsx")
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
x = data.drop(columns=["MEDV"])
y = data["MEDV"]
```

```python
indices_train = np.random.choice(data.index, size=int(0.5 * len(data)), replace=False)
indices_test = data.index.difference(indices_train)
data_train = data.loc[indices_train]
data_test = data.loc[indices_test]
```

```python
x_train = data_train.drop(columns=["MEDV"])
y_train = data_train["MEDV"]

x_test = data_test.drop(columns=["MEDV"])
y_test = data_test["MEDV"]
```

```python
scaler = StandardScaler().fit(x_train)
x_scaled = scaler.transform(x_train)
```

```python
ols = sm.OLS(y_train, sm.add_constant(x_scaled))
results = ols.fit()
results.summary()
```

```python
x_test_scaled = scaler.transform(x_test)
y_pred = results.predict(sm.add_constant(x_test_scaled))
r2 = r2_score(y_test, y_pred)
r2
```

```python
lr = LinearRegression()
lr.fit(x_scaled, y_train)
y_pred_lr = lr.predict(x_test_scaled)
r2_lr = r2_score(y_test, y_pred_lr)  
r2_lr
```

```python
lr_test = LinearRegression()
lr_test.fit(x_test_scaled,y_test)
y_pred_test = lr_test.predict(x_test_scaled)
r2_test = r2_score(y_test, y_pred_test)
r2_test
```
