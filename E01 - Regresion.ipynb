Encontrar la parábola que pasa exactamente por los puntos (1, 2), (1,1) y (3,2). Utiliza la librería sklearn.

```python
x = np.array([1,
              2,
              3])
y = np.array([2,
              1,
              2])
```

```python
x_2 = np.vstack((x, x**2)).T

lr = LinearRegression()
lr.fit(x_2, y)
```

```python
b0, b1, b2 = lr.intercept_, lr.coef_[0], lr.coef_[1]
print(f"Parábola: y = {b0:.2f} + {b1:.2f}x + {b2:.2f}x^2")
```

```python
data = pd.read_csv("imdb_top1000_lae.csv")
data = data.drop(columns = ["Unnamed: 0", 'Series_Title', 'Director', 'Runtime', 'Gross'])
data = data.dropna()
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data["Released_Year"] = pd.to_numeric(data["Released_Year"], errors='coerce')
```

```python
data_dummies = pd.get_dummies(data, columns = ['Certificate'] )
data_dummies = data_dummies.dropna()
```

```python
x = data_dummies.drop(columns=['IMDB_Rating'])
y = data_dummies['IMDB_Rating']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
data_scaled = pd.DataFrame(data_scaled, columns = x_train.columns)
data_scaled.head()
```

```python
model = LinearRegression()
model.fit(data_scaled, y_train) 
y_pred_train = model.predict(data_scaled)
r2_train = r2_score(y_train, y_pred_train)
y_pred_test = model.predict(x_test_scaled)
r2_test = r2_score(y_test, y_pred_test)
```

```python
r2_test, r2_train
```

```python
ols = sm.OLS(y_train.reset_index(drop=True), sm.add_constant(data_scaled.reset_index(drop=True)))
results = ols.fit()
results.summary()
```

P values mayores a 0.05: Todos aquellos parámetros cuyos p-values sean vayores a 0.05, significan que su relación con la variable que se esta pronóticando no existe, aunque pueda parecer que sí, por lo que no se tomarán en cuenta estos factores.



El $R^2$ de test da 0.06069586168259844, lo que indica que el modelo no es bueno para predecir la variable objetivo en datos futuros, ya que solo se está explicando el 6% de la varianza del IMBD rating. El $R^2$ de train da 0.15381606165551287, lo que indica que el modelo no tiene un ajuste moderado a los datos de entrenamiento tampoco, pero no es suficiente para garantizar una buena capacidad de generalización.
