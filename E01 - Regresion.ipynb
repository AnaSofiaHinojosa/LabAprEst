3.        ¿En qué consiste el proceso de escalamiento de factores?

El escalamiento de factores consiste en transformar los valores de las variables para que todas estén en un mismo rango o intervalo de valores. De esta manera, ninguna variable tiene más peso que otra solo por el tamaño de sus valores, y además permite compararlas de forma sencilla y mejora el desempeño de los modelos que usan los datos.

4.        Explica el propósito de penalizar factores en una regresión.
   
El propósito de penalizar factores en una regresión es evitar que los coeficientes (las betas) se hagan demasiado grandes, ya sea reduciéndolos o incluso haciéndolos cero. Esto sirve para que el modelo no dependa solo de unas pocas variables, sino que tome en cuenta la información de manera más equilibrada, evitando que sólo se tome en cuenta cierta información en lugar de todos los factores.

5.        ¿Cuál es la relación entre escalamiento y penalización?

La relación entre escalamiento y penalización es que el escalamiento hace que las betas sean comparables entre sí. De esta manera, si algunas betas crecen demasiado, la penalización puede actuar correctamente. Si no se hace escalamiento, los coeficientes podrían ser grandes no porque el modelo dependa demasiado de esas variables, sino simplemente porque los datos no están en la misma escala, lo que distorsiona la interpretación y la penalización.

6.        Explica el concepto de una prueba de hipótesis.

Una prueba de hipótesis consiste en plantear una hipótesis nula, que generalmente dice que no hay efecto, que no existe relación entre variables, o que algo es igual a un valor específico. La hipótesis nula representa la situación que asumimos verdadera mientras no tengamos evidencia suficiente en contra. Por otro lado, la hipótesis alterna es la afirmación opuesta, se acepta si los datos muestran que la hipótesis nula probablemente no es correcta.

7.        Explica la interpretación de un p-value de una prueba de hipótesis que compara contra una media µ.

El p-value en una prueba de hipótesis que compara contra una media indica la probabilidad de obtener un resultado igual al planteado por la hipótesis nula, por ejemplo, que la beta sea igual a cero, lo que significaría que no hay relación entre la variable y la variable que se quiere predecir, asumiendo que la hipótesis nula es cierta. Mientras más pequeño sea el p-value, menos compatibles son los datos con la hipótesis nula, lo que sugiere que podría existir una relación real entre las variables.

Por ejemplo, cuando queremos ver si una variable es significativa en un modelo, la hipótesis nula dice que no hay relación entre esa variable y lo que queremos predecir, aunque pueda parecer que sí, mientras que la hipótesis alterna dice que sí existe esa relación. Si usamos un nivel de confianza del 95 por ciento, que corresponde a un nivel de significancia de 0.05, un p-value menor a 0.05 indica que los datos son poco compatibles con la hipótesis nula, por lo que se rechaza y se concluye que la variable tiene un efecto significativo.

8.        Describe el propósito de realizar cross-validation.

El propósito de realizar cross-validation es evitar que el modelo se sobreajuste a los datos que se han dado. Esto se hace separando los datos en conjuntos de entrenamiento y prueba, de manera que el modelo se entrene con los datos de entrenamiento y luego se evalúe con los datos de prueba. Así podemos verificar si el modelo realmente generaliza y funciona bien con datos nuevos, o si está sobreajustado y no será capaz de predecir correctamente información que no haya visto antes.

```python
data = pd.read_csv("imdb_top1000_lae.csv")
data = data.drop(columns = ["Unnamed: 0", 'Series_Title', 'Director'])
data = data.dropna()
data.head()
```

```python
data["Released_Year"] = pd.to_numeric(data["Released_Year"], errors='coerce')
```

```python
data['Runtime'] = data['Runtime'].astype(str)
data['Runtime'] = data['Runtime'].str.replace(' min', '').astype(int)

data['Gross'] = data['Gross'].astype(str)
data['Gross'] = data['Gross'].str.replace(',', '').astype(float)
```

```python
data
```

```python
data_dummies = pd.get_dummies(data, columns = ['Certificate'] )
data_dummies = data_dummies.dropna()
```

```python
x = data_dummies.drop(columns=['IMDB_Rating'])
y = data_dummies['IMDB_Rating']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
data_scaled = pd.DataFrame(data_scaled, columns = x_train.columns)
data_scaled.head()
```

```python
data_scaled = data_scaled.to_numpy()
```

```python
model = LinearRegression()
model.fit(data_scaled, y_train) 
y_pred_train = model.predict(data_scaled)
r2_train = r2_score(y_train, y_pred_train)
y_pred_test = model.predict(x_test_scaled)
r2_test = r2_score(y_test, y_pred_test)
```
