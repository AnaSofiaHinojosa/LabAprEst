```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

### Ejercicio 1.1

#### a) Calcula el R2 e interpreta los signos de los betas.

```python
data = pd.read_excel("Motor Trend Car Road Tests.xlsx")
```

```python
data = data.drop(columns = ["model"])
x = data.drop(columns = ["mpg"])
y = data["mpg"]
```

```python
def original(x, y):

    
    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(x)

    lr = LinearRegression()
    lr.fit(x_scaled, y)
    y_pred = lr.predict(x_scaled)
    r2 = r2_score(y, y_pred)
    
    coef_df = pd.DataFrame([lr.coef_], columns=x.columns)
    coef_df['R2'] = r2
    return coef_df
```

```python
a1 = original(x, y)
a1
```

#### b) Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.

```python
def split(x, y, train_size):


    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = train_size)

    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)

    lrb = LinearRegression()
    lrb.fit(x_train, y_train)
    y_pred = lrb.predict(x_test)
    y_pred_train = lrb.predict(x_train)
    r2_train = r2_score(y_train, y_pred_train)
    r2b = r2_score(y_test, y_pred)
    
    tablab = pd.DataFrame([lrb.coef_], columns = x.columns)
    tablab['R2 Train'] = r2_train
    tablab['R2 Test'] = r2b
    return tablab
```

```python
b1 = split(x, y, 0.4)
b1
```

#### c) Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.

```python
def ridge(x, y, alpha):


    l2 = Ridge(alpha = alpha)
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.4)

    l2.fit(x_train, y_train)
    y_pred = l2.predict(x_test)
    y_pred_train = l2.predict(x_train)
    r2_train = r2_score(y_train, y_pred_train)
    r2b = r2_score(y_test, y_pred)

    tablac = pd.DataFrame([l2.coef_], columns = x.columns)
    tablac['R2 Train'] = r2_train
    tablac['R2 Test'] = r2b
    return tablac
```

```python
c1 = ridge(x, y, 1.9)
c1
```
