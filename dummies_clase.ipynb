```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("aguas_feo.xlsx", sheet_name="Hoja3")
```

```python
data_dummies = pd.get_dummies(data, columns=['sabor'])
```

```python
data_dummies
```

```python
import statsmodels.api as sm
x = data_dummies.drop(columns=["cal"])
y = data_dummies["cal"]
x_numeric = x.astype(int)
hstack = np.hstack((np.ones((x_numeric.shape[0], 1)), x_numeric))
ols = sm.OLS(y, hstack)
results = ols.fit()
results.summary()
```
